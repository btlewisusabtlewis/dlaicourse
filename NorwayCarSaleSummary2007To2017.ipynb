```
import pandas as pd
import altair as alt
```

```
url = 'https://raw.githubusercontent.com/btlewisusabtlewis/sample_data/master/norway_new_car_sales_by_month.csv'
sales = pd.read_csv(url)
```

```
sales.head()
```

```
sales.tail()
```

```
sales.info()
```

```
alt.Chart(sales).mark_bar().encode(
    alt.X('Year:N', title='Year'),
    alt.Y('Quantity:Q', title='Total Sales 2007-2017')
)
```

```
df = sales
subs = {'Quantity_Hybrid': 0, 'Quantity_Electric': 0}
df.fillna(value=subs)
```

```
new_sales = df.groupby(by='Year').agg({'Quantity': 'sum',
                              'Quantity_Hybrid': 'sum',
                              'Quantity_Electric': 'sum'}).reset_index()
```

```
new_sales['Hybrid_Pct'] = new_sales['Quantity_Hybrid']/new_sales['Quantity']
```

```
new_sales.head()
```

```
new_sales['Electric_Pct'] = new_sales['Quantity_Electric']/new_sales['Quantity']
```

```
new_sales.head()
```

```
new_sales.tail(15)
```

```
new_sales['Electric_Pct'] = 100.0 * new_sales['Electric_Pct']
```

```
new_sales['Hybrid_Pct'] = 100.0 * new_sales['Hybrid_Pct']
```

```
new_sales.tail(15)
```

```
new_sales['Other_Pct'] = 100.0 - (new_sales['Electric_Pct'] + new_sales['Hybrid_Pct'])
new_sales['Green_Pct'] = new_sales['Electric_Pct'] + new_sales['Hybrid_Pct']
```

```
new_sales.tail(15)
```

```
new_sales.head(15)
```

```
new_sales.drop(columns=['Other'])
```

```
electric_pct = alt.Chart(new_sales, title="Electric Cars Sold in Norway 2007-2017").mark_bar().encode(
    alt.X('Year:O', title = 'Year'),
    alt.Y('Electric_Pct:Q', title = 'Electric Car Percentage'),
    color=alt.value('green')
).properties(
    width=400,
    height=400
)
```

```
hybrid_pct = alt.Chart(new_sales, title="Hybrid Cars Sold in Norway 2007-2017").mark_bar().encode(
    alt.X('Year:O', title = 'Year'),
    alt.Y('Hybrid_Pct:Q', title = 'Hybrid Car Percentage'),
    color=alt.value('blue')
).properties(
    width=400,
    height=400
)
```

```
other_pct = alt.Chart(new_sales, title="Other Cars Sold in Norway 2007-2017").mark_bar().encode(
    alt.X('Year:O', title = 'Year'),
    alt.Y('Other_Pct:Q', title = 'Other Car Percentage'),
    color=alt.value('gray')
).properties(
    width=400,
    height=400
)
```
